```python
# Correção e explicação do código acima:

def selection_sort(arr):
    for i in range (len(arr)-1):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr       
```

## Dois erros foram localizados e corrigidos no código, sendo eles:
### 1. O ponto de inicialização do índice de menor valor ->  para o selection sort, o ponto de start deve ser o próprio índice 'i', já que o menor valor já pode começar nele e 'i + 1' não iria localizá-lo;
### 2. A troca de valores -> essa alteração deve ser feita de forma simultânea 'arr[i], arr[min_idx] = arr[min_idx], arr[i]', já que na forma que estava sendo feita 'arr[min_idx] = arr[i]' e 'arr[i] = arr[min_idx]', os valores seriam substituídos e "perdidos" após a troca.

```python
# Gere uma lista aleatória com 7 números utilizando a biblioteca random:

import random

def selection_sort(arr):
    for i in range(len(arr) - 1):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

lista_ordenada = random.sample(range(1, 50), 7)

print(f"Lista ordenada: {selection_sort(lista_ordenada)}.")
```

### O índice mínimo, ao ser localizado na lista sortida, passa a se tornar o índice 0 e o mesmo processo é praticado para que os demais números menores se enquadrem nos índices adiante (de forma crescente).

```python
# Imprima o maior elemento da lista ordenada SEM usar max() ou sorted():

import random

def selection_sort(arr):
    for i in range(len(arr) - 1):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def find_max(arr):
    maior = arr[0]
    for num in arr:
        if num > maior:
            maior = num
    return maior

lista_ordenada = random.sample(range(1, 50), 7)

print(f"Lista ordenada: {selection_sort(lista_ordenada)}.")
print(f"O maior número encontrado na lista é: {find_max(lista_ordenada)}.")
```
